```python
for model, file in zip(models, files):
    print(file+ "\t", model)
```

The above models were selected using the `auto_arima` method from `pmdarima` which selects the model that minimizes the [Akaike information criterion (AIC)](https://en.wikipedia.org/wiki/Akaike_information_criterion).

A (p,1,q) ARIMA is of the form $$z_t = \alpha_1 z_{t-1} + \alpha_2 z_{t-2} + \cdots + \alpha_p z_{t-p} + \epsilon_t + \beta_1 \epsilon_{t-1} + \cdots + \beta_q \epsilon_{t-q}$$
where $z_t = y_t - y_{t-1}$.

A (0,1,0) ARIMA is of the form $y_t = y_{t-1} + \epsilon_t$ where $\epsilon_t \sim N(0,\sigma^2)$

Notice that a model of this form has zero predictive power for buying or shorting stocks for short term gain as it predicts tomorrow's price to be equal to today's.
