```python
import pandas as pd
import os
from datetime import datetime, timedelta
import datetime as dt
import numpy as np
from finvader import finvader as fv
```

```python
df_articles = pd.read_csv('data/outdated/data_finvader.csv')
df_articles.drop(columns = ['Unnamed: 0.1','Unnamed: 0'], inplace = True)
df_articles['Date & Time'] = pd.to_datetime(df_articles['Date & Time'], utc = True)
df_articles.info()
```

```python
#article dataset had a duplicate 
check = df_articles.duplicated(keep = False)
check[check == True]
df_articles.iloc[23124], df_articles.iloc[23123]
df_articles = df_articles.drop_duplicates()
df_articles.info()
```

```python
def fin_sentiment_score(headline:str, abstract:str, version:str):
    txt = headline + ' ' + abstract
    return fv(text = txt, use_henry= True, use_sentibignomics=True, indicator= version)
```

```python
df_articles['finvader_neg'] = df_articles.apply(lambda x: fin_sentiment_score(x['Headline'], x['Text'], 'neg'), axis = 1)
df_articles['finvader_neu'] = df_articles.apply(lambda x: fin_sentiment_score(x['Headline'], x['Text'], 'neu'), axis = 1)
df_articles['finvader_pos'] = df_articles.apply(lambda x: fin_sentiment_score(x['Headline'], x['Text'], 'pos'), axis = 1)
df_articles.info()
```

```python
df_articles.rename(columns = {'finvader':'finvader_tot'}, inplace = True)
df_articles
```

```python
tech_tickers = ["MSFT", "AAPL", "NVDA", "GOOGL", "AMZN"]
healthcare_tickers = ["LLY", "UNH", "JNJ", "MRK", "ABBV"]
finance_tickers = ["JPM", "V", "MA", "BAC", "WFC"]

def map_sector(ticker):
    if ticker in tech_tickers:
        return "Technology"
    elif ticker in healthcare_tickers:
        return "Healthcare"
    elif ticker in finance_tickers:
        return "Finance"
    else:
        return "Unknown"

df_articles['Sector'] = df_articles['Ticker'].apply(map_sector)
df_articles['Sector'].unique()
```

```python
stock_files = {
    "MSFT": "Microsoft_stock.csv",
    "AAPL": "Apple_stock.csv",
    "NVDA": "Nvidia_stock.csv",
    "GOOGL":"Google_stock.csv",
    "AMZN": "Amazon_stock.csv",
    "JPM":  "JP_Morgan_stock.csv",
    "V":    "Visa_stock.csv",
    "MA":   "Mastercard_stock.csv",
    "BAC":  "Bank_of_America_stock.csv",
    "WFC":  "Wells_Fargo_stock.csv",
    "LLY":  "Eli_Lily.csv",
    "UNH":  "UnitedHealth.csv",
    "JNJ":  "Johnson_and_Johnson_stock.csv",
    "MRK":  "Merck.csv",
    "ABBV": "AbbVie.csv"
}
master_stock_df = pd.DataFrame()
for ticker, file in stock_files.items():    
    stock_df = pd.read_csv('Stock_data/'+file)
    stock_df['Date'] = pd.to_datetime(stock_df['Date'], utc = True)
    stock_df.rename(columns={'Date':'Date & Time'}, inplace=True)
    stock_df['Ticker'] = ticker
    master_stock_df = pd.concat([master_stock_df, stock_df], ignore_index=True)

master_stock_df.info()

    
```

# Manipulating Time

Want to merge the stock and sentiment dataframes, but do it along the stock opening and closing times. Thankfully, everthing has been converted to a standard UTC-0 time, although we should note that the stock times currently in the master_stock_df are NOT the stock market opening times, but rather midnight of the index date in EST/EDT. 

```python
#don't run this multiple times in a row, or you keep shifting the time repeatedly
master_stock_df['Opening Time'] = master_stock_df['Date & Time'] + timedelta(hours = 9, minutes = 30)
master_stock_df['Closing Time'] = master_stock_df['Date & Time'] + timedelta(hours = 16)
master_stock_df.rename(columns = {'Date & Time':'Market Date'}, inplace = True)
master_stock_df
```

```python
open_market_days = master_stock_df['Opening Time'].unique()
closing_market_days = master_stock_df['Closing Time'].unique()
strict_dates = master_stock_df['Date & Time'].unique()
len(open_market_days), len(closing_market_days), len(strict_dates)
```

Now let's adjust the time of our articles to match the next market day by closing time and opening times. 

```python
df_articles['Date & Time']= df_articles['Date & Time'].sort_values()
open_day = pd.DataFrame(open_market_days, columns= ['Next Open Time'])
close_day = pd.DataFrame(closing_market_days, columns= ['Next Closing Time'])
dates = pd.DataFrame(strict_dates, columns = ['Next Market Date'])
df_int1 = pd.merge_asof(df_articles, open_day, right_on = "Next Open Time", left_on='Date & Time', direction = 'forward')

df_int2 = pd.merge_asof(df_articles, close_day, right_on = 'Next Closing Time', left_on = 'Date & Time', direction = 'forward')
df_int3 = pd.merge_asof(df_articles, dates, right_on = 'Next Market Date', left_on = 'Date & Time',direction = 'forward' )
df_int3
len(df_int1), len(df_int2), len(df_int3)
```

```python
common = list(df_articles)
df_a = pd.merge(df_int1, df_int2, on= common)
master_articles_df = pd.merge(df_a, df_int3, on = common)
master_articles_df.info()
```

```python
#save vanilla sentiment scores for posterity
vanilla_vader_scores = master_articles_df[['sentiment_neg', 'sentiment_neu', 'sentiment_pos', 'sentiment_tot', 'URL']]
```

```python
master_articles_df.drop(columns = ['sentiment_neg', 'sentiment_neu', 'sentiment_pos', 'sentiment_tot'], inplace = True)
master_articles_df.info()
```

```python
master_articles_df.rename(columns = {'Date & Time': 'Publishing Time'}, inplace = True)
master_articles_df.drop(columns = ['Date', 'Time'], inplace = True)
master_articles_df.info()
```

# MERGING

We have two dataframes, master_stock_df and master_articles_df. We can merge them in three different ways:

1. Associate an article with the next open time
2. Associate article with next closing time
3. Associate article with next market date

All are valid considerations. For reference, complete.csv currently uses option 2. 

In the following, we will go with option 1. Changes to other options is done by pairing the obvious columns in each dataframe with each other in the merge
