```python
import pandas as pd
import os
import json
import warnings
warnings.filterwarnings('ignore')
```

```python
current_dir = os.getcwd()
news_dir = os.path.join(current_dir, 'data')
news_csv = os.path.join(news_dir, 'news_data_2019_2024_sorted.csv')
news_df = pd.read_csv(news_csv)
```

```python
news_df['Date & Time'] = pd.to_datetime(news_df['Date & Time'], utc=True)
news_df['Date & Time']
```

```python
news_df['Date'] = news_df['Date & Time'].dt.date
news_df['Time'] = news_df['Date & Time'].dt.time
```

```python
news_df
```

```python
tech_tickers = ["MSFT", "AAPL", "NVDA", "GOOGL", "AMZN"]
healthcare_tickers = ["LLY", "UNH", "JNJ", "MRK", "ABBV"]
finance_tickers = ["JPM", "V", "MA", "BAC", "WFC"]

def map_sector(ticker):
    if ticker in tech_tickers:
        return "Technology"
    elif ticker in healthcare_tickers:
        return "Healthcare"
    elif ticker in finance_tickers:
        return "Finance"
    else:
        return "Unknown"
```

```python
news_df['Sector'] = news_df['Ticker'].apply(map_sector)
```

```python
reorder_columns = [
    "Date & Time",
    "Date",
    "Time",
    "Ticker",
    "Sector",
    "Source",
    "Negative Sentiment Score",
    "Neutral Sentiment Score",
    "Positive Sentiment Score",
    "Total Sentiment Score (Compound)",
    "Headline",
    "Text",
    "URL"
]

news_df = news_df[reorder_columns]
```

```python
news_df.rename(columns={'Negative Sentiment Score': 'sentiment_neg'}, inplace=True)
news_df.rename(columns={'Neutral Sentiment Score': 'sentiment_neu'}, inplace=True)
news_df.rename(columns={'Positive Sentiment Score': 'sentiment_pos'}, inplace=True)
news_df.rename(columns={'Total Sentiment Score (Compound)': 'sentiment_tot'}, inplace=True)
```

```python
#news_df 
```

```python
current_dir = os.getcwd()
data_dir = os.path.join(current_dir, 'data')
if not os.path.exists(data_dir):
    os.makedirs(data_dir)

file_path = os.path.join(data_dir, 'news_data_2019_2024_complete.csv')
news_df.to_csv(file_path, index=True)
```

```python
stock_files = {
    "MSFT": "Microsoft_stock.csv",
    "AAPL": "Apple_stock.csv",
    "NVDA": "Nvidia_stock.csv",
    "GOOGL":"Google_stock.csv",
    "AMZN": "Amazon_stock.csv",
    "JPM":  "JP_Morgan_stock.csv",
    "V":    "Visa_stock.csv",
    "MA":   "Mastercard_stock.csv",
    "BAC":  "Bank_of_America_stock.csv",
    "WFC":  "Wells_Fargo_stock.csv",
    "LLY":  "Eli_Lily.csv",
    "UNH":  "UnitedHealth.csv",
    "JNJ":  "Johnson_and_Johnson_stock.csv",
    "MRK":  "Merck.csv",
    "ABBV": "AbbVie.csv"
}

stock_data = {}
stock_dir = os.path.join(current_dir, 'Stock_data')
for ticker, file in stock_files.items():
    stock_csv = os.path.join(stock_dir, file)
    stock_df = pd.read_csv(stock_csv)
    stock_df.rename(columns={'Date':'Date & Time'}, inplace=True)
    stock_df['Date & Time'] = pd.to_datetime(stock_df['Date & Time'], utc=True)
    stock_df['Date'] = stock_df['Date & Time'].dt.date
    stock_df['Time'] = stock_df['Date & Time'].dt.time
    stock_data[ticker] = stock_df

stock_data
```

```python
#make new dataframe, and copy over the Date Time, Ticker, Sector, 
daily_sentiment = news_df.groupby(['Date', 'Ticker']).agg({
    'sentiment_neg': 'mean',
    'sentiment_neu': 'mean',
    'sentiment_pos': 'mean',
    'sentiment_tot': 'mean'
}).reset_index()
```

```python
# Create a master date range DataFrame
start_date = pd.to_datetime('2019-03-15')
end_date = pd.to_datetime('2024-03-15')
date_range = pd.date_range(start=start_date, end=end_date).date
date_range
```

```python
master_df = pd.DataFrame([(date, ticker) for date in date_range for ticker in stock_files.keys()], columns=['Date', 'Ticker'])
```

```python
#master_df
```

```python
merged_df = pd.merge(master_df, daily_sentiment, on=['Date', 'Ticker'], how='right')
```

```python
#merged_df
```

```python
final_df = pd.DataFrame()

for ticker, stock_df in stock_data.items():
    # Filter the merged DataFrame for the current ticker
    merged_df_ticker = merged_df[merged_df['Ticker'] == ticker]

    # Ensure Date columns are in datetime format
    merged_df_ticker['Date'] = pd.to_datetime(merged_df_ticker['Date'])
    stock_df['Date'] = pd.to_datetime(stock_df['Date'])

    # Merge with the stock data
    merged_df_ticker = pd.merge(merged_df_ticker, stock_df[['Date', 'Open', 'High', 'Low', 'Close', 'Volume']], on='Date', how='left')
    merged_df_ticker = merged_df_ticker.sort_values(by='Date')
    merged_df_ticker = merged_df_ticker.ffill().bfill()  # Forward and backward fill missing data

    final_df = pd.concat([final_df, merged_df_ticker], ignore_index=True)

#print(final_df.head())
    
```

```python
import numpy as np 
final_df['sentiment_neg'] = final_df['sentiment_neg'].astype(float).apply(lambda x: np.format_float_positional(x, precision=10))
final_df['sentiment_neu'] = final_df['sentiment_neg'].astype(float).apply(lambda x: np.format_float_positional(x, precision=10))
final_df['sentiment_pos'] = final_df['sentiment_neg'].astype(float).apply(lambda x: np.format_float_positional(x, precision=10))
final_df['sentiment_tot'] = final_df['sentiment_neg'].astype(float).apply(lambda x: np.format_float_positional(x, precision=10))
final_df['Open'] = final_df['Open'].astype(float).apply(lambda x: np.format_float_positional(x, precision=3))
final_df['High'] = final_df['High'].astype(float).apply(lambda x: np.format_float_positional(x, precision=3))
final_df['Low'] = final_df['Low'].astype(float).apply(lambda x: np.format_float_positional(x, precision=3))
final_df['Close'] = final_df['Close'].astype(float).apply(lambda x: np.format_float_positional(x, precision=3))
final_df['Volume'] = final_df['Volume'].astype(float)

#print(final_df.head())
```

```python
final_df['Sector'] = final_df['Ticker'].apply(map_sector)
reorder_columns = [
    "Date",
    "Ticker",
    "Sector",
    "sentiment_neg",
    "sentiment_neu",
    "sentiment_pos",
    "sentiment_tot",
    "Open",
    "High",
    "Low",
    "Close",
    "Volume"
]

final_df = final_df[reorder_columns]
```

```python
current_dir = os.getcwd()
data_dir = os.path.join(current_dir, 'data')
if not os.path.exists(data_dir):
    os.makedirs(data_dir)

file_path = os.path.join(data_dir, 'data_merged.csv')
final_df.to_csv(file_path, index=True)
```
